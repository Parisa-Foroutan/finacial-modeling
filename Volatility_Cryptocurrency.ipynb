# Volatility of Cryptocurrencies

## Data Preparation
Daily Data of Most traded Cryptocurrencies:

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#setting figure size
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 20,10

import datetime as dt
import os
import pandas_datareader.data as web

tickers= ['USDT-USD', 'BTC-USD', 'ETH-USD', 'XRP-USD', 'LTC-USD', 'BCH-USD', 'EOS-USD', 'LINK-USD', 'ADA-USD', 'XMR-USD']
start= dt.datetime(2018,12,31)
end= dt.datetime(2020,12,31)

def compile(tickers, start, end):
  main_df=pd.DataFrame()
  for count,ticker in enumerate(tickers):
    df= web.get_data_yahoo(ticker, start, end, interval='d')
    if not os.path.exists('Cryptocurrencies'):
        os.makedirs('Cryptocurrencies')
    df.to_csv('Cryptocurrencies/{}.csv'.format(ticker))

df = compile(tickers, start, end)
```

### Close price of cryptocurrencies

```
for count,ticker in enumerate(tickers):
  if ticker=='USDT-USD':
    data = pd.read_csv('Cryptocurrencies/{}.csv'.format(ticker), index_col='Date', parse_dates=True)
    data.drop(['High','Low','Open','Adj Close','Volume'], axis=1, inplace= True)
    data.rename(columns={'Close': ticker}, inplace=True)
  else:
    data_count = pd.read_csv('Cryptocurrencies/{}.csv'.format(ticker), index_col='Date', parse_dates=True)
    data_count.drop(['High','Low','Open','Adj Close','Volume'], axis=1, inplace= True)
    data_count.rename(columns={'Close': ticker}, inplace=True)
    data = pd.concat([data, data_count], axis=1)
data.to_csv('Cryptocurrencies/data_Close.csv')
data.head()
```

### Volume and Close Cryptocurrencies

```
for count,ticker in enumerate(tickers):
  if ticker=='USDT-USD':
    data1 = pd.read_csv('Cryptocurrencies/{}.csv'.format(ticker), index_col='Date', parse_dates=True)
    data1.drop(['High','Low','Open','Adj Close'], axis=1, inplace= True)
    data1.rename(columns={'Close': ticker, 'Volume':'Vol_{}'.format(ticker)}, inplace=True)
  else:
    data1_count = pd.read_csv('Cryptocurrencies/{}.csv'.format(ticker), index_col='Date', parse_dates=True)
    data1_count.drop(['High','Low','Open','Adj Close'], axis=1, inplace= True)
    data1_count.rename(columns={'Close': ticker, 'Volume':'Vol_{}'.format(ticker)}, inplace=True)
    data1 = pd.concat([data1, data1_count], axis=1)
data1.to_csv('Cryptocurrencies/data_Vol_Close.csv')
data1.head()
```

### Calculate Return

```
df_USDT = pd.read_csv('Cryptocurrencies/USDT-USD.csv', index_col='Date', parse_dates=True)
df_USDT.drop(['High','Low','Open', 'Close'], axis=1, inplace= True)
df_USDT.head()
```

```
df_USDT.tail()
```

```
df_USDT_returns = 100*(np.log(df_USDT['Adj Close']).pct_change())
df_USDT_returns = pd.DataFrame(df_USDT_returns)
df_USDT_returns.rename(columns={'Adj Close': 'lreturn_pct'},inplace= True)
df_USDT_returns['Volume'] = df_USDT['Volume']
USDT = df_USDT_returns.iloc[1:-1].to_csv('Cryptocurrencies/USDT.csv')
df_USDT_returns.iloc[1:-1].tail(6)
```
